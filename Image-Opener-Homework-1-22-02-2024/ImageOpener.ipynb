```python
from google.colab import files
from google.colab.patches import cv2_imshow
import gif2numpy as gf
import cv2 as cv

def show_image(file_name):
    #get the uploaded file name and ensure it's an image with the correct extension
    if not file_name.endswith(('.png', '.jpg', '.jpeg', '.bmp')):
        print(f"Error: Unsupported image format. Please upload a PNG, JPEG, BMP, or GIF image.")
        exit()

    #read the image using OpenCV
    image = cv.imread(file_name)

    #check if image is read successfully
    if image is None:
        print(f"Error: Could not read image {file_name}.")
        exit()

    #display the image
    cv2_imshow(image)
    cv.waitKey(0)
    cv.destroyAllWindows()
    print("Image opened successfully!")

def show_video(file_name):
    #get the uploaded file name and ensure it's an image with the correct extension
    if not file_name.endswith(('.gif')):
        print(f"Error: Unsupported image format. Please upload a PNG, JPEG, BMP, or GIF image.")
        exit()

    #read the video using OpenCV and Gif2Numpy
    frames, exts, image_specs = gf.convert(file_name)

    #display the video
    for i in range(len(frames)):
        cv2_imshow(frames[i])
        cv.waitKey(0)
    cv.destroyAllWindows()

    print("GIF opened successfully!")
```

```python
#upload an image: PNG
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
show_image(file_name)
```

```python
#upload an image: JPG
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
show_image(file_name)
```

```python
#upload an image: BMP
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
show_image(file_name)
```

```python
#upload an image: GIF
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
show_video(file_name)
```
